### 2.	Write a python program to display IMDB's Top rated 100 movies' data (i.e. Name, IMDB rating, Year of release).

```python
"""
Checking for Web Site Permission, Whether Website allows to perform or to do web scrapping or Not.
"""
import requests
from bs4 import BeautifulSoup 
r1=requests.get("https://www.imdb.com/search/title/?groups=top_100&ref_=adv_prv")
r1.status_code
```

```python
import pandas as pd
import numpy as np
import requests
from requests import get
from bs4 import BeautifulSoup
```

```python
"""
There are two URL used to fetch Top 100 IMDB Movies the first URL will fetch 50 Movies and Second URL will fetch remaining 50 movies
Hence Two Independet Program is written Separately where final o/p of two program is combined as one ouput as list to display 
top 100 IMDB best Movies

1. First URL for top 50 IMDB Movies: url1 = "https://www.imdb.com/search/title/?groups=top_100&ref_=adv_prv/"
2. Next URL for remaining 50 Movies : url2 = "https://www.imdb.com/search/title/?groups=top_100&start=51&ref_=adv_nxt"

"""
url1 = "https://www.imdb.com/search/title/?groups=top_100&ref_=adv_prv/" # This URL Will Display First 50's movies List 
url2 = "https://www.imdb.com/search/title/?groups=top_100&start=51&ref_=adv_nxt" # This is the next URL showing next 50 movies list
headers = {"Accept-Language": "en-US, en;q=0.5"}
results1 = requests.get(url1, headers=headers)
results2 = requests.get(url2, headers=headers)
```

```python
soup1 = BeautifulSoup(results1.text, "html.parser")
soup2 = BeautifulSoup(results2.text, "html.parser")
```

```python
movie_div1 = soup1.find_all('div', class_='lister-item mode-advanced')
movie_div2 = soup2.find_all('div', class_='lister-item mode-advanced')
```

```python
titles1 = []
years1 = []
imdb_ratings1 = []
```

```python
#our loop through each container
for container in movie_div1:

        #name
        name = container.h3.a.text
        titles1.append(name)
        
        #year
        year = container.h3.find('span', class_='lister-item-year').text
        years1.append(year)

        # runtime
       # runtime = container.p.find('span', class_='runtime').text if container.p.find('span', class_='runtime').text else '-'
       # time.append(runtime)

        #IMDb rating
        imdb = float(container.strong.text)
        imdb_ratings1.append(imdb)        
```

```python
titles2 = []
years2 = []
imdb_ratings2 = []
```

```python
for container in movie_div2:

        #name
        name = container.h3.a.text
        titles2.append(name)
        
        #year
        year = container.h3.find('span', class_='lister-item-year').text
        years2.append(year)

        # runtime
       # runtime = container.p.find('span', class_='runtime').text if container.p.find('span', class_='runtime').text else '-'
       # time.append(runtime)

        #IMDb rating
        imdb = float(container.strong.text)
        imdb_ratings2.append(imdb)      
```

```python
titles = titles1 + titles2
years = years1 + years2
imdb_ratings = imdb_ratings1 + imdb_ratings2
```

```python
print(len(titles),len(years),len(imdb_ratings))
```

```python
#pandas dataframe        
movies = pd.DataFrame({
'movie': titles,
'year': years,
'imdb': imdb_ratings,
})
```

### 5. i)	Top 10 ODI teams in men' s cricket along with the records for matches, points andrating.

```python
from bs4 import BeautifulSoup
import requests
from prettytable import PrettyTable
```

```python
def Menu():
    print('\n1. Men \n')
    gen=Gender()
    print('Team Rankings')
    tp=TeamOrPlayer()

    mode=''
    val=''

    if gen=='mens':
        print('ODI')
        mode=Mode()


    return gen,tp,mode,val
```

```python
def Gender():
    
    gender=input('Enter your choice:')
    code={'1':'mens'}
    

    if gender in code:
        return code[gender]
    else:
        
        print('\nInvalid Input\nTry Again\n')
        return Gender();
```

```python
def TeamOrPlayer():
    choice=input('Enter your choice:')
    tp={'1':'team-rankings'}

    if choice in tp:
        return tp[choice]

    else:
        print('\nInvalid Input\nTry Again\n')
        return TeamOrPlayer();
```

```python
def Mode():
    choice=input('Enter your choice:')
    word={'1':'/odi'}

    if choice in word:
        return word[choice]

    else:
        print('\nInvalid Input\nTry Again\n')
        return Choice();
```

```python
def Mode2():
    choice=input('Enter your choice:')
    word={'1':'/odi'}

    if choice in word:
        return word[choice]

    else:
        print('\nInvalid Input\nTry Again\n')
        return Choice();
```

```python
def URL():
    gen,tp,mode,val=Menu()
    url='https://www.icc-cricket.com/rankings/'+gen+'/'+tp+mode+'/'+val
    header=gen.upper() +' ' +mode[1:].upper() + ' ' + val.upper()
    print('\n{:<15}  {:<30}\n{:<15}  {:<30}'.format('',tp.upper(),'',header))
    return url,tp
```

```python
def SOUP(url,tp):
    try:
        res=requests.get(url)
        soup=BeautifulSoup(res.text,'lxml')
        a= soup.find_all('tr',{'class':'table-body'})
        data={}

        for i in a :
            team=[]
            name=''
            rating=''

            try:
                rank=int(i.contents[1].text)
            except:
                pass

            try:
                name=i.contents[3].text.replace('\n','')
                name=" ".join(name.split())
                if rank==1 and tp=='player-rankings':
                    name=name[0:-3]
            except:
                pass

            try:
                rating=i.contents[9].text
            except:
                if rank==1 :
                    rating=i.contents[5].text
                else:
                    rating=i.contents[7].text

            team.extend([name,rating])
            data[rank]=team

        return data

    except:
        return SOUP(url,tp)


def Print(data):
    print('\nRANKING \t TEAM\t\t\t\tRATING')
    for i in sorted(data):
        print('{:<10}       '.format(i),end='')
        for j in range(len(data[i])):
            print('{:<26}'.format(data[i][j]),end='     ')
        print()


def main():

    url,tp=URL()
    data=SOUP(url,tp)
    Print(data)	


if __name__=='__main__':
    main()
```

```python
# to read multiple page of web
def get_url(page_no):
    url = 'https://www.amazon.in/s?bbn=1805560031&rh=n%3A1805560031%2Cp_36%3A1318506031&dc&qid=1632484946&rnid=1318502031&ref=lp_1805560031_nr_p_36_1{}'.format(page_no)
    #url = 'https://www.amazon.in/s?i=electronics&bbn=1805560031&rh=n%3A1805560031%2Cp_36%3A1000000-2000000&dc&qid=1632486672&rnid=1318502031&ref=sr_nr_p_36_2{}'.format(page_no)
    return url

# collect all information
def gather_information():

    # create a dict
    data = {'Description':[],'global_rating':[],'price':[],'rate':[]}
    # just keep loop to read all page
    for pg in range(1,18):
        res = requests.get(get_url(pg))
        markup = res.content
        soup = BeautifulSoup(markup,'lxml')
        phone = soup.find_all('div',class_='s-asin')
        for p in phone:
            try:
                data['Description'].append(p.find('span',class_='a-size-base-plus').text)
            except:
                pass
            data['global_rating'].append(p.find('span',class_='a-size-base').text)
            try:
                data['price'].append(p.find('span',class_='a-price-whole').text)
            except:
                pass
            try:
                data['rate'].append(p.find('span',class_='a-icon-alt').text)
            except:
                data['rate'].append('None')
            
            
    # Now store dataframe as csv file        
    df = pd.DataFrame(data)
    df.to_csv('mobile12.csv')
if __name__ == "__main__":
    gather_information()
```
