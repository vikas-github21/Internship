```python
driver.get(url)
```

```python
job_titles=[]
company_names=[]
locations_list=[]
experience_list=[]
```

```python
try:
    titles_tags=driver.find_elements_by_xpath("//a[@class='title fw500 ellipsis']")
except:
    titles_tags=np.nan
titles_tags[0:10]
```

```python
for i in titles_tags:
    title=i.text
    job_titles.append(title)
job_titles[0:10]
```

```python
for i in companies_tags:
    company_name=i.text
    company_names.append(company_name)
company_names[0:10]
```

```python
try:
    experience_tags=driver.find_elements_by_xpath("//li[@class='fleft grey-text br2 placeHolderLi experience']")
except:
    experience_tags=np.nan
experience_tags[0:10]
```

```python
for i in experience_tags:
    experience=i.text
    experience_list.append(experience)
experience_list[0:10]
```

```python
try:
    locations_tags=driver.find_elements_by_xpath("//li[@class='fleft grey-text br2 placeHolderLi location']/span[1]")
except:
    locations_tags=np.nan
locations_tags[0:10]
```

```python
for i in locations_tags:
    location=i.text
    locations_list.append(location)
locations_list[0:10]
```

```python
print(len(job_titles),len(company_names),len(experience_list),len(locations_list))
```

```python
jobs=pd.DataFrame({})
jobs['title']=job_titles[0:20]
jobs['company']=company_names[0:20]
jobs['experience_required']=experience_list[0:20]
jobs['location']=locations_list[0:20]
```
